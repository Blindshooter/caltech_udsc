```python
%%time
train_metadata = []

for root, dirs, files in os.walk("./train"):
    for sf in dirs:
        if sf.split('.')[0][-1]=='0':
            print('reading files from {}...'.format(sf))
        for f_name in os.listdir(os.path.join("./train", sf)):
            if not f_name.startswith('.'):
                temp = misc.imread(os.path.join("./train", sf, f_name)) # read image
                # collect image metadata
                image_metadata = []
                image_metadata.extend([sf, f_name, os.path.join("./train", sf, f_name)])
                image_metadata.extend( list(temp.shape) if len(temp.shape) == 3\
                                      else [temp.shape[0], temp.shape[1], 1])
                image_metadata.extend([temp.nbytes, temp.dtype])
                # append image metadata to list
                train_metadata.append(image_metadata)
```

```python
train_metadata = pd.DataFrame(train_metadata)
#train_metadata.head()
train_metadata.columns = ['category','img_name', 'full_path', 'height','width','channels','byte_size','bit_depth']
# select integer target (1-257)
train_metadata['target'] = train_metadata['category'].str.split('.').apply(lambda x: int(x[0]))
# image extension
train_metadata['img_extension'] = train_metadata['img_name'].str.split('.').apply(lambda x: x[1])
# just name without category indexes
train_metadata['category_name'] = train_metadata['category'].str.split('.').apply(lambda x: x[1]).str.lower()
```

# Train image resize

```python
def image_preproc_save(path_to_file, path_to_save, resize_dim = (40,40)):
    image = misc.imread(path_to_file)
    if len(image.shape) == 2:
        image = np.stack((image,)*3)
    misc.imsave(path_to_save,misc.imresize(image, size=resize_dim))
```

```python
%%time
for root, dirs, files in os.walk("./train"):
    for sf in dirs:
        if sf.split('.')[0][-1]=='0':
            print('Reading files from {}...'.format(sf))
        for f_name in os.listdir(os.path.join(root, sf)):
            if not f_name.startswith('.'):
                load_path = os.path.join(root, sf, f_name)
                save_path = os.path.join(root+"_transformed", sf, f_name)
                try:
                    image_preproc_save(load_path, save_path, (64,64))
                except:
                    print("Failed_preprocessing for: {}".format(load_path))
```

```python
from keras.utils import np_utils
from sklearn.datasets import load_files

data = load_files('train_transformed')
caltech_files = np.array(data['filenames'])
caltech_targets = np.array(data['target'])

#def load_dataset(path):
#    data = load_files(path)
#    caltech_files = np.array(data['filenames'])
#    caltech_targets = np_utils.to_categorical(np.array(data['target']), 257)
#    return caltech_files, caltech_targets

#train_files, train_targets = load_dataset('train')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(caltech_files, caltech_targets, test_size=0.30, random_state=242, stratify = caltech_targets)
if len(set(y_train))==len(set(y_val)):
    y_train = np_utils.to_categorical(y_train, 257)
    y_val = np_utils.to_categorical(y_val, 257)
    print("Split successful")
else:
    print("Split unsuccessful, unique train classes: {}, unique val classes: {}".format(len(set(y_train)),len(set(y_val))))
```

```python
CHANNELS = 3
ROWS = 64
COLS = 64


def prep_data(images):
    count = len(images)
    data = np.ndarray((count,  ROWS, COLS, CHANNELS), dtype=np.uint8)

    for i, image_file in enumerate(images):
        try:
            img = misc.imread(image_file)
            data[i] = img#.T#.reshape(ROWS, COLS, CHANNELS)
            
        except:
            print("Unable to load "+image_file)
        if i%1000 == 0: print('Processed {} of {}'.format(i, count))
    
    return data
```

```python
print("Train processing")
train = prep_data(X_train)
print ("Shape train "+str(train.shape))
print("-----------------")
print ("Validation processing")
val = prep_data(X_val)
print ("Shape val "+ str(val.shape))
```

```python
from keras.layers import Conv2D, MaxPooling2D, GlobalAveragePooling2D
from keras.layers import Dropout, Flatten, Dense
from keras.models import Sequential

model = Sequential()

model.add(Conv2D(filters=16, kernel_size=2, padding='same', activation='relu', 
                        input_shape=(224, 224, 3)))
model.add(MaxPooling2D(pool_size=2))
model.add(Conv2D(filters=32, kernel_size=2, padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=2))
model.add(Conv2D(filters=64, kernel_size=2, padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=2))
model.add(Dropout(0.3))
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(133, activation='softmax'))

model.summary()
```

```python
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
from keras.callbacks import ModelCheckpoint, ReduceLROnPlateau, EarlyStopping  

epochs = 20

checkpointer = ModelCheckpoint(filepath='saved_models/weights.best.from_scratch.hdf5', 
                               verbose=1, save_best_only=True)
model.fit(X_train, y_train, 
          validation_data=(X_val, y_val),
          epochs=epochs, batch_size=20, callbacks=[checkpointer], verbose=1)
```
